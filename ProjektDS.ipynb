# RFM

```python
data['orderdate'] = pd.to_datetime(data['orderdate'])
```

```python
NOW = dt.datetime(2016,1,1)
rfmTable = data.groupby('householdid').agg({'orderdate': lambda x: (NOW - x.max()).days, # Recency
                                        'orderid': lambda x: len(x),      # Frequency
                                        'totalprice': lambda x: x.sum()}) # Monetary Value

rfmTable['orderdate'] = rfmTable['orderdate'].astype(int)
rfmTable.rename(columns={'orderdate': 'recency', 
                         'orderid': 'frequency', 
                         'totalprice': 'monetary_value'}, inplace=True)
```

```python
if rfmTable['recency'] <= 380:
    rfmTable['Break point'].append(1)
elif rfmTable['recency'] <= 615:
    rfmTable['Break point'].append(2)
elif rfmTable['recency'] <= 1067:
    rfmTable['Break point'].append(3)
elif rfmTable['recency'] <= 1686:
    rfmTable['Break point'].append(4)
else :
    rfmTable['Break point'].append(5)
```
